```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, Button, HBox, VBox, Output
from IPython.display import display, clear_output
```

```python
# PATHS
# ==========================
mr_path = "/content/drive/MyDrive/pat59_cropped.nii.gz"  # Your mr
seg_folder = "/content/TotalSegmentator/segmentations"                # Folder where heart .nii.gz files exist

# ==========================
# LOAD mr
# ==========================
mr_img = nib.load(mr_path)
mr_data = mr_img.get_fdata()
print("mr shape:", mr_data.shape)

# ==========================
# HEART PARTS EXPEmrED
# ==========================
heart_parts = {
    "Heart": "heart.nii.gz",
    "Myocardium": "heart_myocardium.nii.gz",
    "Left Atrium": "heart_atrium_left.nii.gz",
    "Right Atrium": "heart_atrium_right.nii.gz",
    "Left Ventricle": "heart_ventricle_left.nii.gz",
    "Right Ventricle": "heart_ventricle_right.nii.gz",
    "Aorta": "aorta.nii.gz",
    "Pulmonary Artery": "pulmonary_artery.nii.gz"
}

# ==========================
# LOAD MASKS
# ==========================
masks = {}
for label, fname in heart_parts.items():
    path = os.path.join(seg_folder, fname)
    if os.path.exists(path):
        masks[label] = nib.load(path).get_fdata()
        print(f" Loaded: {fname}")
    else:
        print(f" Missing: {fname}")

# ==========================
# COLOR MAPS (RGBA)
# ==========================
colors = {
    "Heart": (1, 0, 0, 0.4),           # Red
    "Myocardium": (0.9, 0.5, 0, 0.5),  # Orange
    "Left Atrium": (0, 0, 1, 0.4),     # Blue
    "Right Atrium": (0, 1, 0, 0.4),    # Green
    "Left Ventricle": (1, 1, 0, 0.4),  # Yellow
    "Right Ventricle": (1, 0, 1, 0.4),  # Magenta
    "Aorta": (1, 0, 0, 0.5),
    "Pulmonary Artery": (0, 1, 1, 0.5)
}

max_slices = mr_data.shape[2] - 1

# ==========================
# VIEWER FUNCTION
# ==========================
def view_slice(slice_idx):
    """Display mr with heart overlays"""
    fig, ax = plt.subplots(figsize=(12, 10))

    ax.imshow(mr_data[:, :, slice_idx].T, cmap="gray", origin="lower")

    for label, mask in masks.items():
        masked_data = np.ma.masked_where(mask[:, :, slice_idx].T == 0, mask[:, :, slice_idx].T)
        ax.imshow(
            masked_data,
            cmap=plt.cm.colors.ListedColormap([colors[label]]),
            origin="lower",
            alpha=colors[label][3]
        )

    ax.set_title(f"Heart Segmentation - Slice {slice_idx}/{max_slices}", fontsize=16, fontweight='bold', pad=20)
    ax.axis("off")

    # Add legend
    legend_elements = [
        plt.Rectangle((0, 0), 1, 1, fc=colors[label][:3], alpha=colors[label][3], label=label)
        for label in masks.keys()
    ]
    if legend_elements:
        ax.legend(handles=legend_elements, loc='upper right', fontsize=10)

    plt.tight_layout()
    plt.show()

# ==========================
# INTERACTIVE SLIDER
# ==========================
print("\n" + "=" * 70)
print("mr HEART VIEWER - Google Colab Version")
print("=" * 70)
print("\nCONTROLS:")
print("  Use the slider to move through slices")
print("  Drag the slider or click on the track")
print("=" * 70)
print(f"\nTotal slices: {max_slices + 1} (0 to {max_slices})")
print("\nHeart parts:")
for label in heart_parts.keys():
    print(f"  - {label}")
print("=" * 70 + "\n")

interact(
    view_slice,
    slice_idx=IntSlider(
        min=0,
        max=max_slices,
        step=1,
        value=max_slices // 2,
        description='Slice:',
        continuous_update=False,
        layout={'width': '600px'},
        style={'description_width': '60px'}
    )
)
```
